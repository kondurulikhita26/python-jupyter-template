### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

# Define x values from 0 to 10
x = np.arange(0, 11)
# Define y values as the square of x
y = x ** 2

# Create the plot
plt.plot(x, y, marker='o', linestyle='-', color='b', label='y = x^2')

# Customizations
plt.title('Line Plot Customization')  # Adding a title
plt.xlabel('X Values')  # Label for X-axis
plt.ylabel('Y Values (X Squared)')  # Label for Y-axis
plt.grid(True)  # Adding a grid
plt.legend()  # Adding a legend

# Display the plot
plt.show()
```
